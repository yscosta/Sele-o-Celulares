## Seleção de celular pelo método Electre I

### Caso 1

#### O objetivo dessa aplicação é a utilização do método Electre I para selecionar celulares entre 4 marcas.

#### Iniciou-se o processo selecionando as marcas que atendem os critérios eliminatórios para as 4 marcas que são a memória de 4GB/128GB, bateria de 5000 mAh e câmera de 50 megapixel.

#### No passo seguinte foram adotados os critérios de minimização de preço, minimização de espessura do celular, maximização do tamanho da tela e minimização do peso.

#### Os dados foram coletados e organizados na planilha "Seleção de Celular.xlsx" onde será adicionada a coluna da recomendação pelo Electre I em relação das marcas que superam as demais.

#### Os requisitos técnicos para executar esse notebook pelo jupyter onde previamente deverá ter os pacotes pandas e numpy instalados no ambiente do interpretador Python. Na mesma pasta desse notebook deverão estar o arquivo "Seleção de Celular.xlsx" e o pacote ElectreyFamily.

```python
from ElectreFamily import ElectreI
```

```python
electreI_c1 = ElectreI('Seleção de Celular.xlsx')
```

```python
electreI_c1.agreement_threshold = 0.8
electreI_c1.disagreement_threshold = 0.2

electreI_c1.exec()
electreI_c1.model.to_excel('Seleção de Celular - Recomendado.xlsx')
electreI_c1.model
```

### Caso 2

#### Nessa aplicação foi baixado uma base de dados do Kaggle sobre celulares da Ucrânia.

https://www.kaggle.com/datasets/artempozdniakov/ukrainian-market-mobile-phones-data

#### A base de dados foi ajustada removendo as instância que tinham característica com dados faltantes. Foram preservadas as colunas popularity, best_price, sellers_amount, screen_size, memory_size, battery_size e release_date, sendo que essa última foi transformada para numérico seguindo a máscara ano e mês, ou seja, aaaamm01.

#### Foram adicionados as linhas de objetivos entre minimização e maximização, assim como, os pesos de importância.

```python
electreI_c2 = ElectreI('Ukraine_phones_data.xlsx')
```

```python
electreI_c2.agreement_threshold = 0.8
electreI_c2.disagreement_threshold = 0.2

electreI_c2.exec()
electreI_c2.model.groupby('Recomendação')['Recomendação'].count()
```
